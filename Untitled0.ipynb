Q1. What is the difference between a function and a method in Python?
- Functions = A block of code that is defined independently and can be called anywhere. The Functions are not bound to any object. It is Called directly using its name. It can be Can be standalone or part of a module.
For Example:-
```
def add(a, b):
    """A function to add two numbers."""
    return a + b
#Calling the function
result = add(10, 5)
print(f"Result from function: {result}")
#Output = Result from function: 15
```

- Method = A block of code associated with an object and called using the object. The Methods are bound to an object or class. Called using an object or class (object.method() or class.method()). The Method is Always part of a class or an instance. For Example:-
```
class Calculator:
    def add(self, a, b):
        """A method to add two numbers."""
        return a + b
# Creating an object of the class
calc = Calculator()
# Calling the method using the object
result = calc.add(10, 5)
print(f"Result from method: {result}")
```

Q5. What are iterators in Python and how do they differ from iterables?
- In Python, iterables and iterators are two closely related concepts, but they serve different purposes when working with loops and data processing. Here's an explanation of each and their differences:-

1. An iterable is any Python object capable of returning its members one at a time. These objects implement the __iter__() method or the __getitem__() method. An iterable can be looped over using a for loop or be passed to functions like iter(). For Example:-
```
# A list is an iterable
my_list = [1, 2, 3]
# You can iterate over it using a for loop
for item in my_list:
    print(item)
```
2. Iterator = An iterator is an object that represents a stream of data. It is the result of calling iter() on an iterable. Iterators are objects that implement two key methods: __iter__() which returns the iterator object itself. __next__() which returns the next item from the container and raises a StopIteration exception when no more items are available. For Example:-
```
# Get an iterator from the list
my_list = [1, 2, 3]
iterator = iter(my_list)
# Use the iterator to fetch elements one by one
print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2
print(next(iterator))  # Output: 3
# If you call next() again, it will raise StopIteration
# print(next(iterator))  # Raises StopIteration
```





Q6. Explain the concept of generators in Python and how they are defined ?
- A generator is a special type of iterator in Python. It is a function that produces a sequence of values, but unlike a regular function that returns a single value, a generator can yield multiple values, one at a time, using the yield keyword. Generators are memory-efficient because they generate values on the fly, rather than storing them in memory all at once.
- Generators are defined like normal functions, but instead of returning a value using return, they use the yield statement. For Example:-
```
def simple_generator():
    yield 1
    yield 2
    yield 3
# Create a generator object
gen = simple_generator()
# Using next() to get values one by one
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
```



 Q7. What are the advantages of using generators over regular functions ?
 - Using generators instead of regular functions in Python offers several advantages, especially in scenarios involving large datasets or complex iterative processes. Here are the key benefits of using generators:-

1. Memory Efficiency = Generators are memory-efficient because they do not store the entire sequence of values in memory. Instead, they generate values one at a time when needed. For Example:-
```
def generate_numbers(limit):
    count = 1
    while count <= limit:
        yield count
        count += 1
# Creating a generator for 1 million numbers
numbers = generate_numbers(1000000)
```
2. Lazy Evaluation = Generators produce values lazily, meaning they compute each value only when it's requested. This can be particularly useful when the entire sequence might not need to be evaluated immediately or fully. For Example:-
```
def read_large_file(filename):
    with open(filename) as file:
        for line in file:
            yield line
```
3. Improved Performance with Large Data = enerators are faster than using lists or other data structures for processing large sequences because they avoid the overhead of constructing and holding the entire list in memory. For Example:-
```
# Generator example that computes Fibonacci numbers lazily
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b
```
4. Infinite Sequences = Generators are ideal for working with infinite sequences. Since they generate values on demand, they can theoretically produce an infinite number of items, without ever requiring the entire sequence to be stored. For Example:-
```
def infinite_fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b
# Get the first 10 Fibonacci numbers
fib_gen = infinite_fibonacci()
for _ in range(10):
    print(next(fib_gen))
```
5. Cleaner and More Readable Code = Generators simplify complex loops or data transformations. When you use a generator, the code tends to be more concise and easier to read compared to manually managing state in traditional functions. For Example:-
```
def squared_numbers(nums):
    for num in nums:
        yield num * num
```

Q10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- map() = The map() function in Python applies a specified function to each item in an iterable (or multiple iterables) and returns an iterator with the results. It is used to transform each element of an iterable independently, based on the function provided. The function is applied to each item sequentially, and the transformed items are yielded one by one. For Example:-
```
# Function to square numbers
def square(x):
    return x * x
# List of numbers
numbers = [1, 2, 3, 4, 5]
# Using map to apply the square function to each element
squared_numbers = map(square, numbers)
# Converting map object to list and printing
print(list(squared_numbers))  # Output: [1, 4, 9, 16, 25]
```
- reduce() = The reduce() function, found in the functools module, takes a binary function (a function that operates on two arguments) and applies it cumulatively to the items of an iterable, from left to right, to reduce the iterable to a single value. It is typically used for operations like summing all elements, multiplying elements together, or finding the greatest common divisor, where a combination of results over multiple items is needed. For Example:-
```
from functools import reduce
# Function to add two numbers
def add(x, y):
    return x + y
# List of numbers
numbers = [1, 2, 3, 4, 5]
# Using reduce to calculate the sum of all numbers
total = reduce(add, numbers)
# Printing the result
print(total)  # Output: 15
```
- filter() = The filter() function applies a function to each item in an iterable and returns an iterator containing only the items for which the function returns True. It is used to filter elements from an iterable based on a given condition, effectively removing elements that do not satisfy the condition specified by the function. For Example:-
```
# Function to check if a number is even
def is_even(x):
    return x % 2 == 0
# List of numbers
numbers = [1, 2, 3, 4, 5, 6]
# Using filter to get even numbers
even_numbers = filter(is_even, numbers)
# Converting filter object to list and printing
print(list(even_numbers))  # Output: [2, 4, 6]
```

````python
# Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num, "is prime")
````

````python
# Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.next_value = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.terms:
            current_value = self.current
            self.current, self.next_value = self.next_value, self.current + self.next_value
            self.count += 1
            return current_value
        else:
            raise StopIteration

fib = FibonacciIterator(10)
for num in fib:
    print(num)
````

````python
# Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_2(5):
    print(power)
````

````python
# Q7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_lines(file_path):
    """Reads a file line by line and yields each line as a string.

    Args:
        file_path (str): The path to the file to read.

    Yields:
        str: Each line of the file.
    """
    try:
        with open(file_path, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: File not found at path: {file_path}")
        return

file_path = r'C:\Users\panch\OneDrive\Desktop\PwSkills\Jupyter\example.txt'

import os
if os.path.exists(file_path):
    for line in read_file_lines(file_path):
        print(line)
else:
    print(f"Error: File not found at path: {file_path}")
````

````python
# Q9.  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.


def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius_temperatures = [0, 20, 30, 40, 100]

fahrenheit_temperatures = map(celsius_to_fahrenheit, celsius_temperatures)

print(list(fahrenheit_temperatures))
````

````python
# Q10. Create a Python program that uses `filter()` to remove all the vowels from a given string.


def is_not_vowel(char):
    return char.lower() not in 'aeiou'

input_string = "Hello, World!"

filtered_string = ''.join(filter(is_not_vowel, input_string))

print(filtered_string)
````
